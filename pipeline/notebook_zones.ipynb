```python
url = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
url
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql://root:root@localhost:5432/taxi_zone_lookup")
```

```python
df_zones.to_sql(
    name="taxi_zone",
    con=engine,
    if_exists="replace"
)
```
